```python
import numpy as np
import pandas as pd

df1 = pd.read_csv('kaggle_train.csv')
df1.drop('Unnamed: 0', 1, inplace=True)

df2 = pd.read_csv('twitter_dataset.csv', sep=",")

indexNames = df1[ ~(df1['Emoji']==0) & # 0
                 ~(df1['Emoji']==2) & # 1
                 ~(df1['Emoji']==4) & # 2
                 ~(df1['Emoji']==7) & # 3
                 ~(df1['Emoji']==18) & # 4
                 ~(df1['Emoji']==19) ].index #5

df1.drop(indexNames , inplace=True)

df1['Emoji'] = df1['Emoji'].replace([2],1)
df1['Emoji'] = df1['Emoji'].replace([4],2)
df1['Emoji'] = df1['Emoji'].replace([7],3)
df1['Emoji'] = df1['Emoji'].replace([18],4)
df1['Emoji'] = df1['Emoji'].replace([19],5)

data = pd.concat([df2,df1])
data.to_csv("new_dataset.csv", index = False)
data
```
